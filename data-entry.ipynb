```python
for x in json_obj["data"]:
    for skill in x["spells"]:
        skill_url = x["passive"]["image_url"]
        response = requests.get(skill_url)
        if response.status_code == 200:
            with open(f"C:\PLProjects\PyProjects\s5-university-website\media\skills\icon\{x['name']}_Passive.jpg", "wb") as f:
                f.write(response.content)
                print(f"Image downloaded successfully!: {x['name']} ")
        else:
            print(f"Failed to download image. Status code: {response.status_code} Image: {x['name']}")
```
